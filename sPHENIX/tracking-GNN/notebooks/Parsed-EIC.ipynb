```python
# ───────── Plot N random barrel-pixel events (uniform style) ──────────
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import random
plt.style.use('ggplot')

EVENT_DIR = Path("/ssd3/giorgian/eic/events")   # directory containing event_*.npz
N_EVENTS  = 10                      # number of random events to display
```

```python
# ---------------------------------------------------------------------

# collect every per-event file
files = sorted(EVENT_DIR.glob("event_*.npz"))
if not files:
    raise FileNotFoundError(f"No event_*.npz files found in {EVENT_DIR}")

# choose up to N_EVENTS distinct events at random
chosen = random.sample(files, min(N_EVENTS, len(files)))

# prepare grid of sub-plots
ncols = 2
nrows = (len(chosen) + ncols - 1) // ncols
fig, axes = plt.subplots(nrows, ncols, figsize=(6*ncols, 6*nrows),
                         constrained_layout=True)
axes = axes.flat if hasattr(axes, "flat") else [axes]

for ax, f in zip(axes, chosen):
    d = np.load(f)

    x, y, z = d["x_mm_si"], d["y_mm_si"], d["z_mm_si"]
    #print(f'{x.shape=} {y.shape=} {z.shape=}')


    ax.scatter(x, y, s=20, label='SiBarrel')

    x, y, z = d["x_mm_bw"], d["y_mm_bw"], d["z_mm_bw"]
    #print(f'{x.shape=} {y.shape=}')


    ax.scatter(x, y, s=20,  label='SiBackwards')

    x, y, z = d["x_mm_ecal"], d["y_mm_ecal"], d["z_mm_ecal"]

    ax.scatter(x, y, s=20, label='ECal')

    
    x, y, z = d["x_mm_hcal"], d["y_mm_hcal"], d["z_mm_hcal"]

    ax.scatter(x, y, s=20, label='HCal')

    
    ax.set_title(f.stem)           # strip ".npz"
    ax.set_xlabel("x [mm]")
    ax.set_ylabel("y [mm]")
    ax.legend()
    #ax.set_aspect("equal")

# hide any unused sub-plots
for ax in axes[len(chosen):]:
    ax.axis("off")

fig.tight_layout()
plt.show()
```
