```python
def calc_dphi(phi1, phi2):
    """Computes phi2-phi1 given in range [-pi,pi]"""
    dphi = phi2 - phi1
    dphi[dphi > np.pi] -= 2*np.pi
    dphi[dphi < -np.pi] += 2*np.pi
    return dphi
```

```python
def find_medoid(pixels):
    # pixels: (N, 3)
    deltas = pixels[None] - pixels[:, None]
    # deltas: (N, N, 3)
    deltas = np.linalg.norm(deltas, axis=0)
    # deltas: (N, N)
    mean_distance = np.mean(deltas, axis=0)
    return pixels[np.argmin(mean_distance)]
    
```

```python
stds = []
counts = []
is_trigger = []
layers = []
medoid_diffs_raw = []
medoid_diffs = []
clusters = []
clusters_raw = []
momentums = []
for file in tqdm(files[:10000]):
    f = np.load(file)
    pixels = f['pixel_cartesian']
    r = np.sqrt(pixels[:, 0]**2 + pixels[:, 1]**2)
    phi = np.arctan2(pixels[:, 1], pixels[:, 0])
    z = pixels[:, 2]
    cartesian = np.stack([r, z], axis=-1)
    hit_id = f['hit_id']
    trigger_node = f['trigger_node']
    layer_ids = f['layer_id']
    for hit_id in range(len(f['hit_cartesian'])):
        mask = f['hit_id'] == hit_id
        momentums.append(f['momentum'][hit_id])
        n_pixels = np.sum(mask)
        selected_pixels = pixels[mask]
        counts.append(n_pixels)
        layers.append(layer_ids[hit_id])
        clusters_raw.append(np.stack([cartesian[mask][:, 0], phi[mask], cartesian[mask][:, 1]], axis=-1))
        r_std, z_std = np.std(cartesian[mask], axis=0)
        selected_phi = phi[mask]
        x = np.mean(np.cos(selected_phi))
        y = np.mean(np.sin(selected_phi))
        phi_std = np.sqrt(-np.log(min(x**2 + y**2, 1)))
        
        is_trigger.append(trigger_node[hit_id])
        #diff = np.abs(selected_phi - mean_phi)
        #diff = np.minimum(diff, np.abs(2*np.pi - diff))
        #phi_std = np.sqrt(np.sum(diff**2))
        stds.append([r_std, phi_std, z_std])

        selected_pixels = np.array(sorted(tuple(p) for p in selected_pixels))
        medoid = find_medoid(selected_pixels)
        diffs = selected_pixels - medoid[None]
        medoid_diff = np.array(sorted([tuple(p) for p in diffs]))
        medoid_diff_orig = np.array(medoid_diff)
        medoid_diffs_raw.append(medoid_diff_orig)
        medoid_diff = np.abs(medoid_diff)
        medoid_diff[medoid_diff == 0] = np.max(medoid_diff)
        min_deltas = np.min(medoid_diff, axis=0)
        min_deltas[min_deltas == 0] = 1
        medoid_diff = np.round(medoid_diff_orig/min_deltas).astype(int)
        medoid_diffs.append(np.array(medoid_diff))
        min_x, min_y, min_z = np.min(medoid_diff[:, 0]), np.min(medoid_diff[:, 1]), np.min(medoid_diff[:, 2])
        medoid_diff[:, 0] -= min_x
        medoid_diff[:, 1] -= min_y
        medoid_diff[:, 2] -= min_z
        clusters.append(medoid_diff)
        
stds = np.stack(stds, axis=0)
is_trigger = np.array(is_trigger).astype(bool)
counts = np.array(counts)
layers = np.array(layers)
momentums = np.array(momentums)
```

```python
cluster_counts = defaultdict(int)
trigger_cluster_counts = defaultdict(int)
cluster_sigma_phi = {}
for i, cluster in enumerate(tqdm(clusters)):
    cluster = tuple(sorted([tuple(p) for p in cluster]))
    cluster_counts[cluster] += 1
    if is_trigger[i]:
        trigger_cluster_counts[cluster] += 1
    if cluster not in cluster_sigma_phi:
        cluster_sigma_phi[cluster] = stds[i][1]
```
