```python
layer_preds = defaultdict(list)
layer_gts = defaultdict(list)
layer_pairs =  np.array([                                                   
       (0, 6), (1, 6), (2, 6), (3, 6), (4, 6),                                   
       (0, 5), (1, 5), (2, 5), (3, 5), (4, 5),                                   
       (0, 4), (1, 4), (2, 4),                                                   
       (0, 3), (1, 3), (2, 3),                                                   
       (6, 4), (6, 3),                                                           
       (5, 4), (5, 3)                                                            
       ])                                                                        
          
```

```python
for batch in tqdm(islice(valid_data_loader, 0, 10000)):
    edge_index = batch.edge_index.detach().cpu().numpy()
    preds = model(batch.to(DEVICE)).detach().cpu().numpy()
    targets = batch.y.squeeze(0).detach().cpu().numpy()
    layer_ids = batch.x[:, 4]
    for (l1, l2) in layer_pairs:
        mask = (((layer_ids[edge_index[0, :]] == l1) & (layer_ids[edge_index[1, :]] == l2)) | ((layer_ids[edge_index[0, :]] == l2) & (layer_ids[edge_index[1, :]] == l1))).cpu().numpy()
        layer_preds[(l1, l2)].extend(preds[mask])
        layer_gts[(l1, l2)].extend(targets[mask])
        
```

```python
gts
```

```python
recalls = []
precisions = []
accuracies = []
aucs = []
pair_labels = []
for pair in sorted(layer_pairs, key=lambda x: x[1]-x[0]):
    pair = tuple([int(x) for x in pair])
    preds = np.array(layer_preds[pair])
    gts = np.array(layer_gts[pair])
    recall = round(recall_score(gts, preds > 0).item()*100, 1)
    precision = round(precision_score(gts, preds > 0).item()*100, 1)
    accuracy = round(accuracy_score(gts, preds > 0)*100, 1)
    auc = round(roc_auc_score(gts, preds).item()*100, 1)
    #print(f'[Layers {pair}]')
    #print(f'{recall=}\n{precision=}\n{accuracy=}\n{auc=}\n')

    recalls.append(recall)
    precisions.append(precision)
    accuracies.append(accuracy)
    aucs.append(auc)
    pair_labels.append(str(pair))  # convert pair to string for labeling

```
