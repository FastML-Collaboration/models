```python
import numpy as np
import glob
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
```

```python
trigger_files_y = glob.glob('/ssd3/giorgian/hits-data-october-2024/trigger/1/*-y.npz')
trigger_files = glob.glob('/ssd3/giorgian/hits-data-october-2024/trigger/1/*.npz')
trigger_files_z = list(set(trigger_files) - set(trigger_files_y))
```

```python
ips_y = []
for file in tqdm(trigger_files_y):
    f = np.load(file)
    ips_y.append(f['interaction_point'])
ips_y = np.array(ips_y)
```

```python
ips_z = []
for file in tqdm(trigger_files_z):
    f = np.load(file)
    ips_z.append(f['interaction_point'])
ips_z = np.array(ips_z)
```

```python
plt.hist(ips_y[:, 0], color='orange', alpha=0.3, label='y', density=True)
plt.hist(ips_z[:, 0], color='blue', alpha=0.3, label='z', density=True)
plt.legend()
```

```python
plt.hist(ips_y[:, 1], color='orange', alpha=0.3, label='y', density=True)
plt.hist(ips_z[:, 1], color='blue', alpha=0.3, label='z', density=True)
plt.legend()
```
