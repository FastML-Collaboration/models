```python
S = torch.rand((15, 15))
s = S @ S.T
```

```python
import torch
from torch.distributions import MultivariateNormal

# Define the mean vector and covariance matrix
mean = torch.zeros(s.shape[0])  
covariance = s  #

# Create the multivariate normal distribution
distribution = MultivariateNormal(mean, covariance)

# Sample from the distribution
sample = distribution.sample(sample_shape=(100000,))
```

```python
sample.shape[0]
```

```python
torch.cov(sample.T)
```

```python
(torch.abs(torch.cov(sample.T) - s)/s)*100
```
